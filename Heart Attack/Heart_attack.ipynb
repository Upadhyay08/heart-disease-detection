```python
print(df.describe())
```

```python
cat_cols = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal']

fig, axes = plt.subplots(2, 4, figsize=(16, 8))
axes = axes.flatten()

for i, col in enumerate(cat_cols):
    df[col].value_counts().plot(kind='bar', ax=axes[i], color='steelblue')
    axes[i].set_title(col)
    axes[i].set_xlabel('')

plt.tight_layout()
plt.show()
```

```python
print(df.dtypes)
```

```python
num_cols = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']
fig, axes = plt.subplots(1, 5, figsize=(20, 4))
axes = axes.flatten()

for i, col in enumerate(num_cols):
    df[col].plot(kind='hist', ax=axes[i], color='steelblue', bins=20)
    axes[i].set_title(col)

plt.tight_layout()
plt.show()
```

# Bivariate Analysis 

```python
fig, axes = plt.subplots(1, 5, figsize=(20, 5))

for i, col in enumerate(num_cols):
    sns.boxplot(data=df, x='target', y=col, ax=axes[i])
    axes[i].set_title(f'{col} vs target')

plt.tight_layout()
plt.show()
```

### HeatMap to see the correlation between the Features and the Target Variable

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
